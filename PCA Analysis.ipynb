```python
 
import numpy as np
```

```python
x1 = [2.5, 0.5, 2.2, 1.9, 3.1, 2.3, 2, 1, 1.5, 1.1]
 
 
```

```python
mean = np.mean(x1)
std_dev = np.std(x1)
 
standardized_data_x1 = (mean - x1)
standardized_data_x1 = standardized_data_x1
 
print("Standardized data x1:")
print(standardized_data_x1)
```

```python
x2 = [2.4, 0.7, 2.9, 2.2, 3.0, 2.7, 1.6, 1.1, 1.6, 0.9]
 
mean = np.mean(x2)
std_dev = np.std(x2)
 
standardized_data_x2 = (x2 - mean)
standardized_data_x2 = standardized_data_x2
 
print("\nStandardized data x2:")
print(standardized_data_x2)
```

```python
data = np.array([standardized_data_x1, standardized_data_x2])
 
cov_matrix = np.cov(data)
 
print("\nCovariance matrix:")
print(cov_matrix)
```

```python
from numpy.linalg import eig
 
w, v = eig(cov_matrix)
 
print("\nEigenvalues (w):")
print(w)
 
print("\nEigenvectors (v):")
print(v)
 
print("\nnp.transpose(v):")
print(np.transpose(v))
```

```python

a = np.transpose(standardized_data_x1)
b = np.transpose(standardized_data_x2)
 
data_t = np.array([a, b])
 
print("\ndata_t:")
print(data_t)
```

```python
 
final = np.transpose(v) @ data_t
print("\nfinal:")
print(final)
```

```python
print("\nnp.transpose(final):")
print(np.transpose(final))
 
```
