```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Sample data
X = np.array([1, 2, 3, 4, 5 , 7,8, 9 , 10]).reshape(-1, 1)
y = np.array([24, 36, 53, 77, 11, 45, 78, 90, 100])
```

```python
#model
model = LinearRegression()
model.fit(X,y)
```

```python
#prediction
Y_pred = model.predict(X)
```

```python
#output
print("coefficient:", model.coef_[0])
print("intercept:", model.intercept_)
print("predicted values:", Y_pred)
```

```python
#performance
n= len(X)
SSF= mean_squared_error(y, Y_pred)*n
R2 = r2_score(y, Y_pred)
print("Sum of Squared Errors (SSF):", SSF)
print("R2 score:", R2)
```

```python
#sort for smooth line
sorted_indices = np.argsort(X.flatten())
X_sorted = X[sorted_indices]
y_pred_sorted = Y_pred[sorted_indices]
```

```python
#plot
plt.scatter(X, y,  label='Actual Data')
plt.plot(X_sorted, y_pred_sorted, color='red', label='Regression Line')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
